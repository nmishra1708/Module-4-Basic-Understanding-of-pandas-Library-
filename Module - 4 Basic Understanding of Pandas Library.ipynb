# Learn about Merge, Join and Concatenate

```python
# Understanding about Merge Operations
# 1. Merge -> Perform operation using attaributes 
# 2. Join -> Perform operation using index number
# 3. Concate -> Perform on the values
```

```python
# Create two DataFrames
```

```python
# First Data Frame
player = ['player1','player2','player3']
points = [8,9,6]
title = ['game1','game2','game3']
df1 = pd.DataFrame({'Player':player,'Points':points,'Title':title})
df1
```

```python
# Second DataFrame
player = ['player1','player5','player6']
power = ['punch','kick','elbow']
title = ['game1','game5','game6']
df2 = pd.DataFrame({'Player':player,'Power':power,'Title':title})
df2
```

```python
# Perform Merge Operations
```

```python
# 1. Inner Join Operation

# Here three parameters use for perform merge operation 
# 1. parameter dataframe name, 
# 2. parameter on which attribute used for apply merge operation and 
# 3. how to perform operation.
```

```python
df1.merge(df2, on='Title',how='inner')
```

```python
# 2. Left Merge operation
df1.merge(df2, on='Title',how='left')
```

```python
# 3. Right Merge Operation
df1.merge(df2, on='Title',how='right')
```

```python
# 4. Outer Merge Operation
df1.merge(df2, on='Title',how="outer")
```

# Join Operations 

```python
'''
for implement join operation need to put index value on DataFrame for better understanding.
'''
```

```python
# Create DataFrames
player = ['player1','player2','player3']
points = [8,9,6]
title = ['game1','game2','game3']
df3 = pd.DataFrame({'Player':player,'Point':points,'Title':title}, index=['l1','l2','l3'])
df3
```

```python
player = ['player1','player2','player3']
points = [8,9,6]
title = ['game1','game2','game3']
df4 = pd.DataFrame({'Players':player,'Points':points,'Titles':title},index=['l2','l3','l4'])
df4
```

```python
# 1. Inner Join
df3.join(df4, how='inner')
```

```python
# left join
df3.join(df4, how='left')
```

```python
# right join
df3.join(df4, how='right')
```

```python
# outer join
df3.join(df4, how='outer')
```

### Concatenate two DataFrames

```python
pd.concat([df3,df4])
```

# Cleaning/Cleansing  the Dataset

```python
'''
1. Data Cleaning or Data cleansing is the process of detecting, correcting or removing corrupt inaccurate records from a dataset, tables or database.
2. It also refers to identifying incomplete, incorrect, inaccurate or irrelevant parts of the data and then replacing, modifying or deleting the dirty data.
'''
```

```python
cars.head()
```

```python
# Rename columns 
cars = cars.rename(columns={'Unnamed: 1':'model'})
```

```python
cars.head()
```

```python
# Change mpg from string to float
cars.mpg = cars.mpg.astype(float)
```

```python
# Plotting area plot
y1 = cars['hp']
y2 = cars['disp']
x = range(32)
plt.stackplot(x,y1,color='purple',alpha=0.7)
plt.stackplot(x,y2,color='black',alpha=0.5)
```
